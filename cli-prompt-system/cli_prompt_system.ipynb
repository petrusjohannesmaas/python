# Timeblocking CLI tool (Programming Language Capstone)

- To practise basic data types and structures of Python I'm going to build a CLI app to help me manage my work schedule
- I've adjusted my roadmap based on this Youtube video: https://www.youtube.com/watch?v=3QP43F4gVIE

### Project Overview

The primary goals of this app is to encourage healthier programming habits and increase my productivity

**When starting, the user is prompted to:** 
0. STARTUP PROTOCOL
1. Create a research pile for the day (25m)
2. Rehydrate (5m)
3. Study (90m)
4. Eat something in silence (15m)
5. Work (90m)
6. Coffee break (5m)
7. Read documentation (60m)
8. Stretch and eat lunch(30m) #320 
9. Capstone hackathon (180m)
12. Walk and rehydrate (15m)
13. Design patterns (30m)
14. SHUTDOWN PROTOCOL (15m)
